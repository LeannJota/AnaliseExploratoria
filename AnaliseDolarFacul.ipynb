Este código irá gerar três gráficos diferentes: um gráfico de linha mostrando a variação de USD_BRL ao longo do tempo, um histograma mostrando a distribuição de USD_BRL, e um boxplot também mostrando a distribuição de USD_BRL.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Dataset - USD_BRL_hist.csv")
```

```python
df.info()
```

```python
# Convertendo a coluna 'data' para o tipo datetime
df['Data'] = pd.to_datetime(df['Data'])

# Definindo 'data' como índice
df.set_index('Data', inplace=True)
```

```python
# Criando o primeiro gráfico (linha)
plt.figure(figsize=(10,5))
plt.plot(df['USD_BRL'])
plt.title('Gráfico de Linha: USD_BRL ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('USD_BRL')
plt.show()
```

```python
# Criando o segundo gráfico (histograma)
plt.figure(figsize=(10,5))
plt.hist(df['USD_BRL'], bins=30, alpha=0.5, color='g')
plt.title('Histograma: Distribuição de USD_BRL')
plt.xlabel('USD_BRL')
plt.ylabel('Frequência')
plt.show()
```

```python
# Criando o terceiro gráfico (boxplot)
plt.figure(figsize=(10,5))
plt.boxplot(df['USD_BRL'])
plt.title('Boxplot: Distribuição de USD_BRL')
plt.ylabel('USD_BRL')
plt.show()
```
